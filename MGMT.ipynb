```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python


# 数据
years = [2017, 2018, 2019, 2020, 2021]
disney_share = [25, 24, 23, 22, 20]
netflix_share = [30, 32, 35, 37, 40]
amazon_share = [20, 21, 22, 23, 25]

# 绘制图表
plt.figure(figsize=(12, 8))
plt.plot(years, disney_share, marker='o', linestyle='-', color='blue', label='Disney')
plt.plot(years, netflix_share, marker='s', linestyle='-', color='orange', label='Netflix')
plt.plot(years, amazon_share, marker='^', linestyle='-', color='green', label='Amazon')

# 添加标题和标签
plt.xlabel('Year', fontsize=14)
plt.ylabel('Market Share (%)', fontsize=14)
plt.title('Market Share Trends (2017-2021)', fontsize=16)
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

# 显示图表
plt.show()
```

```python


# 数据
years = [2017, 2018, 2019, 2020, 2021]
disney_subs = [30, 40, 50, 60, 80]
netflix_subs = [110, 125, 140, 160, 190]
amazon_subs = [70, 80, 90, 100, 120]

# 绘制图表
plt.figure(figsize=(12, 8))
plt.plot(years, disney_subs, marker='o', linestyle='-', color='blue', label='Disney')
plt.plot(years, netflix_subs, marker='s', linestyle='-', color='orange', label='Netflix')
plt.plot(years, amazon_subs, marker='^', linestyle='-', color='green', label='Amazon')

# 添加标题和标签
plt.xlabel('Year', fontsize=14)
plt.ylabel('Subscribers (in millions)', fontsize=14)
plt.title('Subscriber Growth (2017-2021)', fontsize=16)
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

# 显示图表
plt.show()

```
